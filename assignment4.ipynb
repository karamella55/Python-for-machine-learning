```python
#####task1########

class rec():
    def __init__(self,w,l):
        self.w=w
        self.l=l
    def parameter(self):
        return 2*(self.w+self.l)
    def area(self):
        return self.w*self.l
    def display(self):
        print("width:",self.w)
        print("length:",self.l)
        print("parameter:",self.parameter())
        print("area",self.area())



class Parallelepipede(rec):
    def __init__(self,w,l,h):
        self.h=h
        rec.__init__(self,w,l)
    def volume(self):
        return self.w*self.l*self.h


ree=rec(2,3)
ree.display()
pp=Parallelepipede(2,3,4)
print("volume:",pp.volume())
        
```

```python
########task2########
class person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print("name",self.name)
        print("age",self.age)
class child(person):
    def __init__(self,name,age,section):
        self.section=section
        person.__init__(self,name,age)
    def display(self):
        print("name",self.name)
        print("age",self.age)
        print("section",self.section)

p=person("karamella",20)
p.display()
c=child("karamellax",10,"A")
c.display()
```

```python
def longest_common_subsequence(s1,s2):
    x,y=len(s1),len(s2)
    L=[[0]*(y+1) for i in range(x+1)]
    
    for i in range(1,x+1):
        for j in range(1,y+1):
            if s1[i-1]==s2[j-1]:
                L[i][j]=L[i-1][j-1]+1
            else:
                L[i][j]=max(L[i-1][j],L[i][j-1])
    a,b=x,y
    l2=[]
    while a!=0 and b!=0:
        if L[a][b]==L[a-1][b]:
            a-=1
        elif L[a][b]==L[a][b-1]:
            b-=1
        else:
            l2.append(s1[a-1])
            a-=1
            b-=1
    return "".join(l2[::-1])

st1="karamella"
st2="meanalla"
print(longest_common_subsequence(st1,st2))
```
